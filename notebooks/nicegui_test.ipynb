```python
# Import necessary libraries
import pandas as pd
```

```python
# Load stock data from CSV
stock_data = pd.read_csv('../stock_data.csv', parse_dates=['Date'], dayfirst=True)
```

```python
stock_data.head()
```

```python
# List of companies (based on the CSV columns)
companies = stock_data.columns[1:].tolist()
```

```python
# Test stock data fetching for one company (e.g., TCS.NS)
symbol = 'TCS.NS'
stock_data_filtered = stock_data[['Date', symbol]]
```

```python
stock_data_filtered.head()  # Display the first few rows
```

```python
# Optionally, you can plot the data (if you want to see it visually)
import matplotlib.pyplot as plt
```
